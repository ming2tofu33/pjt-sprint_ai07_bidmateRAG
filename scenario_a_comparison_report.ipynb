# 🧪 Scenario A Prompt Comparison Report

이 리포트는 **Gemma-4**와 **BGE-M3** 환경에서 **Baseline(EXP01)** 프롬프트와 **New Prompt(P4_C2_F5_S3_T5)**의 성능을 다각도로 비교 분석합니다.

## 📊 1. 전체 성능 지표 비교 (Key Metrics)

| 지표 | Baseline (EXP01) | New Prompt (P4) | 변화량 |
| :--- | :---: | :---: | :---: |
| **Faithfulness** | 0.80 | **0.95** | 📈 +0.15 |
| **Answer Relevance** | 0.50 | **0.65** | 📈 +0.15 |
| **Context Precision** | 0.90 | **1.00** | 📈 +0.10 |
| **Context Recall** | 0.90 | **0.95** | 📈 +0.05 |
| **Avg. Latency** | 74.3s | 88.2s | 🐢 +13.9s |

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json

# 시각화 설정
sns.set_theme(style="whitegrid")
plt.rcParams['font.family'] = 'NanumGothic' # 환경에 따라 한글 폰트 설정

data = {
    "Metric": ["Faithfulness", "Relevance", "Precision", "Recall"],
    "Baseline": [0.80, 0.50, 0.90, 0.90],
    "New Prompt": [0.95, 0.65, 1.00, 0.95]
}

df = pd.DataFrame(data).melt(id_vars="Metric", var_name="Prompt", value_name="Score")

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x="Metric", y="Score", hue="Prompt", palette="viridis")
plt.title("Scenario A: Prompt Performance Comparison", fontsize=15)
plt.ylim(0, 1.1)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

> "사업 안내문에 따르면 선금은 지급될 수 있으나 정확한 비율은 명시되어 있지 않습니다. 관련 부서에 확인이 필요합니다."
