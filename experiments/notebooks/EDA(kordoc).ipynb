## pdf / hwp 전부 일관 변환 -> .mb 마크다운 형식으  (텍스트 추출) 

````python
# .hwp 또는 .pdf로 끝나는 파일만 골라서 리스트로 만듦
files = [f for f in os.listdir(DATA_DIR) if f.endswith((".hwp", ".pdf"))]

# 변환 결과 추적용
ok_count = 0
fail_list = []

for i, fname in enumerate(files):
    # 상대경로를 절대경로로 변환 (kordoc이 절대경로를 필요로 함)
    input_path = os.path.abspath(os.path.join(DATA_DIR, fname))
    # 확장자를 .md로 바꿔서 출력 파일명 생성 (예: 벤처기업.hwp → 벤처기업.md)
    output_name = os.path.splitext(fname)[0] + ".md"
    # 출력 폴더 + 출력 파일명 합쳐서 저장 경로 생성
    output_path = os.path.join(OUT_DIR, output_name)

    # npx kordoc 명령어를 실행하여 hwp/pdf → md 변환
    kordoc_cmd = (
        'export NVM_DIR="$HOME/.nvm" && '
        '[ -s "$NVM_DIR/nvm.sh" ] && . "$NVM_DIR/nvm.sh" && '
        "nvm use 20 >/dev/null && "
        f"npx kordoc {shlex.quote(input_path)} -o {shlex.quote(output_path)}"
    )
    result = subprocess.run(
        ["bash", "-lc", kordoc_cmd], capture_output=True, encoding="utf-8", errors="replace"
    )

    # 성공/실패 기록
    if result.returncode == 0:
        ok_count += 1
    else:
        fail_list.append(fname)

    # 진행 상황 출력 (예: [1/100] OK - 벤처기업.hwp)
    status = "OK" if result.returncode == 0 else "FAIL"
    print(f"[{i + 1}/{len(files)}] {status} - {fname}")

# 최종 요약
print(f"\n{'=' * 40}")
print(f"성공: {ok_count}/{len(files)}")
print(f"실패: {len(fail_list)}개")
if fail_list:
    print("실패 파일:")
    for f in fail_list:
        print(f"  - {f}")
````

````python
# 확인
parsed_files = sorted(OUT_DIR.glob("*.md"))
source_files = sorted([p for p in DATA_DIR.iterdir() if p.suffix.lower() in {".hwp", ".pdf"}])

print(f"원본 파일 수: {len(source_files)}")
print(f"변환된 md 파일 수: {len(parsed_files)}")

empty_files = [p.name for p in parsed_files if p.stat().st_size == 0]
print(f"빈 md 파일 수: {len(empty_files)}")
if empty_files:
    print("빈 파일:")
    for name in empty_files[:10]:
        print(f"  - {name}")

parsed_names = {p.stem for p in parsed_files}
source_names = {p.stem for p in source_files}

missing_outputs = sorted(source_names - parsed_names)
extra_outputs = sorted(parsed_names - source_names)

print(f"\n변환 안 된 파일 수: {len(missing_outputs)}")
if missing_outputs:
    for name in missing_outputs[:10]:
        print(f"  - {name}")

print(f"\n원본에는 없는데 parsed에만 있는 파일 수: {len(extra_outputs)}")
if extra_outputs:
    for name in extra_outputs[:10]:
        print(f"  - {name}")
````

````python
# 3. 특수문자/노이즈 확인
noise_pattern = re.compile(r"[^\w\s가-힣ㄱ-ㅎㅏ-ㅣ\.,\(\)\[\]\|·•□◦○※①②③\-\:]")

noise_data = []
for fname in os.listdir(OUT_DIR):
    if not fname.endswith(".md"):
        continue
    path = os.path.join(OUT_DIR, fname)
    with open(path, "r", encoding="utf-8") as f:
        text = f.read()
    noise_chars = noise_pattern.findall(text)
    noise_data.append(
        {"파일명": fname, "노이즈_수": len(noise_chars), "노이즈_샘플": list(set(noise_chars))[:10]}
    )

df_noise = pd.DataFrame(noise_data).sort_values("노이즈_수", ascending=False).reset_index(drop=True)
print(f"노이즈 평균: {df_noise['노이즈_수'].mean():.0f}개")
df_noise.head(10)
````

````python
# 별첨/양식 섹션 — 패턴 및 위치 EDA

# 탐지할 후보 패턴 (발견 목적 — 확정 아님)
APPENDIX_PATTERNS = [
    # 붙임
    r"\[붙임\s*\d*\]",
    r"^\s*붙임\s*\d+",
    r"## 붙임",
    # 별표
    r"\[별표\s*\d*\]",
    r"## \[별표",
    # 별지
    r"\[별지\s*\d*\]",
    r"^\s*별지\s*서식",
    r"## 별지",
    # 별첨
    r"\[별첨\s*\d*\]",
    r"^\s*별첨",
    # 서식 관련 (신규 추가)
    r"제안서\s*관련\s*서식",
    r"【별도서식】",
    r"\[서식\s*제?\s*\d+\s*호?\]",
    r"^\s*붙임\s*서식",
    r"^\s*서식\s*\d+",
    r"^\s*\[서식\]",
    r"별도\s*서식",
    r"서식\s*제\s*\d+\s*호",
    r"\[별지\s*제\s*\d+\s*호\s*서식?\]",
    r"별지\s*제\s*\d+\s*호\s*서식",
]
compiled = [(p, re.compile(p, re.MULTILINE)) for p in APPENDIX_PATTERNS]

rows = []
for fname in os.listdir(OUT_DIR):
    if not fname.endswith(".md"):
        continue
    path = os.path.join(OUT_DIR, fname)
    with open(path, encoding="utf-8") as f:
        text = f.read()
    total_len = len(text)

    hits = []
    for pattern_str, pattern in compiled:
        for m in pattern.finditer(text):
            hits.append(
                {
                    "패턴": pattern_str,
                    "매칭텍스트": m.group()[:50],
                    "위치_char": m.start(),
                    "위치_%": round(m.start() / total_len * 100, 1),
                }
            )

    rows.append(
        {
            "파일명": fname,
            "총길이": total_len,
            "별첨_히트수": len(hits),
            "첫_별첨_위치_%": min((h["위치_%"] for h in hits), default=None),
            "발견_패턴목록": list({h["패턴"] for h in hits}),
            "히트_상세": hits,
        }
    )

df_appendix = pd.DataFrame(rows).sort_values("첫_별첨_위치_%")

# 요약
print("=== 별첨/서식 패턴 발견 현황 ===")
print(f"전체 파일 수: {len(df_appendix)}")
print(f"패턴 있는 파일: {(df_appendix['별첨_히트수'] > 0).sum()}")
print(f"패턴 없는 파일: {(df_appendix['별첨_히트수'] == 0).sum()}")
print()

# 패턴별 등장 빈도
all_patterns = [p for row in df_appendix["발견_패턴목록"] for p in row]
print("=== 패턴별 등장 파일 수 ===")
print(pd.Series(Counter(all_patterns)).sort_values(ascending=False).to_string())
print()

# 첫 별첨 위치 분포
print("=== 첫 별첨/서식 시작 위치 분포 (%) ===")
print(df_appendix["첫_별첨_위치_%"].describe())
````

## 표 비중 확인 -> 청킹 전략 

````python
# 표 비중 확인
data2 = []
for fname in os.listdir(OUT_DIR):
    if not fname.endswith(".md"):
        continue
    path = os.path.join(OUT_DIR, fname)
    with open(path, "r", encoding="utf-8") as f:
        text = f.read()
    lines = text.splitlines()
    non_empty = [line for line in lines if line.strip()]
    table_lines = [line for line in lines if line.strip().startswith("|")]
    data2.append(
        {
            "파일명": fname,
            "글자수": len(text),
            "표_라인수": len(table_lines),
            "전체_라인수": len(non_empty),
            "표_비중": len(table_lines) / len(non_empty) if non_empty else 0,
        }
    )

df2 = pd.DataFrame(data2).sort_values("표_비중", ascending=False).reset_index(drop=True)
print(f"표 비중 평균: {df2['표_비중'].mean() * 100:.1f}%")
print(f"표 비중 0% 문서: {len(df2[df2['표_비중'] == 0])}개")
df2[["파일명", "글자수", "표_라인수", "전체_라인수", "표_비중"]].head(10)
````

# RAG 시스템 설계 결정을 위한 추가 EDA 방향

현재까지의 EDA를 통해 다음 세 가지는 확인했습니다.
- 문서 길이 편차가 크다.
- 표 비중이 높고, 큰 표도 적지 않다.
- 번호 기반 섹션 구조가 반복적으로 나타난다.

하지만 이것만으로는 실제 RAG 설계 결정을 바로 내리기 어렵습니다. 특히 아래 항목은 추가 확인이 필요합니다.
- 표를 본문과 함께 쓸지, 별도로 분리할지
- 어떤 노이즈를 cleaning rule로 제거할지
- 별첨/부록/서식 구간을 어디서부터 분리할지
- 전처리 이후 본문 기준으로 chunk size와 overlap을 어떻게 정할지

따라서 이후 EDA는 단순 통계 확인이 아니라, 전처리와 retrieval 전략을 정당화하는 방향으로 진행합니다.

## 1. 표 품질 상세 분석

### 왜 필요한가
- 표는 문서 안에서 비중이 큰 정보 단위이지만, 모든 표가 검색에 유용한 것은 아닙니다.
- 특히 레이아웃형 표, 깨진 표, 큰 표는 retrieval과 chunking 품질을 떨어뜨릴 수 있습니다.

### 무엇을 확인하려는가
- Markdown 표 구조가 실제로 유지되는지
- 정보형 표보다 레이아웃형 표가 더 많은지
- 큰 표가 본문 chunk를 방해할 정도로 많은지
- 깨진 행이 많은 문서를 별도 처리해야 하는지

### 이 분석으로 얻고 싶은 결정
- 표를 본문과 같이 넣을지, 분리할지
- 큰 표와 깨진 표를 cleaning 또는 제외할지
- 표 전용 chunking이 필요한지

아래 코드는 `kordoc`로 변환된 Markdown 문서에서 표 품질을 점검하기 위한 EDA입니다.

단순히 표 개수를 세는 것이 아니라, 아래 항목을 함께 봅니다.
- Markdown 표 구조가 유지되는지
- 실제 컬럼형 표처럼 활용 가능한지
- 레이아웃 유지용 표가 많은지
- 빈 셀이 많아 깨진 표가 많은지
- 큰 표가 chunking을 방해할 가능성이 있는지

즉, 이 분석의 목적은 `표를 본문과 같이 RAG에 넣을지`, `표를 별도 정제/분리할지` 결정하기 위한 근거를 만드는 것입니다.

````python
table_quality = []

for fname in os.listdir(OUT_DIR):
    # markdown 파일만 분석
    if not fname.endswith(".md"):
        continue

    path = os.path.join(OUT_DIR, fname)
    with open(path, "r", encoding="utf-8") as f:
        lines = f.readlines()

    tables = []
    current = []

    # 표 후보 추출
    # "|"가 2개 이상 들어간 줄이 2줄 이상 연속되면 표로 간주
    # 본문 문장 속 "|"는 제외하려는 목적
    for line in lines:
        stripped = line.rstrip()

        if stripped.count("|") >= 2:
            current.append(stripped)
        else:
            if len(current) >= 2:
                tables.append(current)
            current = []

    # 마지막 표 처리
    if len(current) >= 2:
        tables.append(current)

    total_tables = len(tables)

    # 표 품질 관련 통계 변수
    markdown_header_tables = 0  # markdown 구분선(|---|)이 있는 표 수
    meaningful_header_tables = 0  # 실제 컬럼 헤더처럼 보이는 표 수
    layout_like_tables = 0  # 레이아웃용 표처럼 보이는 표 수
    broken_rows = 0  # 빈 셀이 과반 이상인 행 수
    large_tables = 0  # 20줄 초과 표 수

    for tbl in tables:
        # markdown 헤더 구분선 여부 확인
        has_md_header = any(
            re.match(r"^\|?\s*:?-{3,}:?\s*(\|\s*:?-{3,}:?\s*)+\|?$", row) for row in tbl
        )
        if has_md_header:
            markdown_header_tables += 1

        # 첫 번째 행을 실제 헤더 후보로 가정
        first_row = tbl[0]
        cells = [c.strip() for c in first_row.strip("|").split("|")]

        # 비어 있지 않은 셀 비율 계산
        non_empty_cells = [c for c in cells if c not in ("", "-", "—")]
        non_empty_ratio = len(non_empty_cells) / len(cells) if cells else 0

        # 의미 있는 헤더 판단
        # - 비어 있지 않은 셀이 2개 이상
        # - 셀 대부분이 비어 있지 않음
        # - 단순 번호/빈칸 위주가 아님
        if (
            len(non_empty_cells) >= 2
            and non_empty_ratio >= 0.6
            and not all(re.match(r"^\d+\.?$", c) or c == "" for c in cells)
        ):
            meaningful_header_tables += 1

        # 레이아웃형 표 판단
        # - 빈 셀이 많거나
        # - 실제 값이 거의 없거나
        # - 표라기보다 정렬용 블록처럼 보이는 경우
        if (
            non_empty_ratio < 0.5
            or len(non_empty_cells) <= 1
            or sum(1 for c in cells if c == "") >= len(cells) / 2
        ):
            layout_like_tables += 1

        # 깨진 행 계산
        # markdown 구분선은 제외
        for row in tbl:
            if re.match(r"^\|?\s*:?-{3,}:?\s*(\|\s*:?-{3,}:?\s*)+\|?$", row):
                continue

            row_cells = [c.strip() for c in row.strip("|").split("|")]
            if not row_cells:
                continue

            empty_cells = sum(1 for c in row_cells if c in ("", "-", "—"))
            if empty_cells / len(row_cells) > 0.5:
                broken_rows += 1

        # 큰 표 집계
        # 표 길이가 너무 길면 chunking을 방해할 수 있음
        if len(tbl) > 20:
            large_tables += 1

    table_quality.append(
        {
            "파일명": fname,
            "총_표수": total_tables,
            "markdown_헤더표수": markdown_header_tables,
            "의미있는_헤더표수": meaningful_header_tables,
            "레이아웃형_표수": layout_like_tables,
            "깨진행수": broken_rows,
            "큰표수(20줄초과)": large_tables,
            "의미있는_헤더비율": round(meaningful_header_tables / total_tables, 2)
            if total_tables
            else 0.0,
            "레이아웃형_표비율": round(layout_like_tables / total_tables, 2)
            if total_tables
            else 0.0,
        }
    )

df_tbl2 = pd.DataFrame(table_quality)
````

## 2. 노이즈 패턴 유형화

표 처리 방향을 본 뒤에는, 본문 안의 노이즈를 어떤 규칙으로 제거할 수 있는지 정리해야 합니다.

### 왜 필요한가
- 문서 안에는 검색에 도움을 주지 않는 반복 기호, 페이지 번호, 헤더/푸터, 일반 계약 문구가 섞여 있습니다.
- 이런 요소를 그대로 두면 retrieval 품질과 답변 집중도가 떨어질 수 있습니다.

### 무엇을 확인하려는가
- 반복 특수문자와 구분선이 얼마나 많은지
- 페이지 번호, 헤더/푸터가 반복되는지
- 여러 문서에 공통으로 등장하는 boilerplate 문구가 있는지

### 이 분석으로 얻고 싶은 결정
- cleaning rule 초안 만들기
- 제거할 노이즈와 유지할 본문 정보를 구분하기

### 예시 패턴
- 반복 특수문자
- 페이지 번호
- 헤더/푸터
- 계약 일반문구
- 불필요한 구분선

````python
# 문서 안에서 반복적으로 등장하는 특수문자 후보를 찾고, 어떤 문서가 상대적으로 특수문자 후보가 많은지 확인하는 EDA
# 한글, 영문, 숫자, 기본 문장부호를 제외한 문자들을 우선 노이즈 후보로 본다.
noise_pattern = re.compile(r"[^\w\s가-힣ㄱ-ㅎㅏ-ㅣ\.,\(\)\[\]\|·•□◦○※①②③\-:%/]")

# 문서별 결과를 저장할 리스트 생성
noise_data = []

# 변환된 markdown 문서가 있는 폴더를 하나씩 순회
for fname in os.listdir(OUT_DIR):
    # markdown 파일만 분석 대상으로 사용
    if not fname.endswith(".md"):
        continue

    # 현재 문서의 전체 경로 생성
    path = os.path.join(OUT_DIR, fname)

    # 문서를 utf-8 인코딩으로 열어서 전체 텍스트를 읽음
    with open(path, "r", encoding="utf-8") as f:
        text = f.read()

    # 현재 문서 안에서 특수문자 후보를 모두 찾음
    candidate_chars = noise_pattern.findall(text)

    # 문서별 요약 정보를 리스트에 저장
    noise_data.append(
        {
            # 어떤 파일에서 나온 결과인지 저장
            "파일명": fname,
            # 특수문자 후보가 총 몇 개 등장했는지 저장
            "특수문자후보_수": len(candidate_chars),
            # 중복 제거 후 대표 문자 샘플만 앞에서 15개 저장
            "특수문자후보_샘플": list(sorted(set(candidate_chars)))[:15],
            # 문서 전체 길이를 저장
            "글자수": len(text),
            # 문서 길이 대비 특수문자 후보 비율 계산
            "특수문자후보_비율": round(len(candidate_chars) / len(text), 4) if len(text) > 0 else 0,
        }
    )

# 리스트를 데이터프레임으로 바꾸고
# 특수문자 후보 개수가 많은 문서 순으로 정렬
df_noise2 = (
    pd.DataFrame(noise_data).sort_values("특수문자후보_수", ascending=False).reset_index(drop=True)
)
````

문맥에 따라 의미가 있을 수 있음 예를 들어
| :표 구조 유지에 필요

①②③ : 문서의 항목 번호일 수 있음

%, :, / :수치, 비율, 날짜, 경로 등에서 필요

※, •, · : 목록 구분 기호일 수 있음

이걸 보는 목적은 이 문자 다 지우자가 아니라 어떤 문자/ 패턴이 많이 나오고, 실제로 제거 가능한지 확인하자 이다

````python
# 전체 요약 보기
# 전체 문서에서 특수문자 후보 개수와 비율이 어떻게 분포하는지 요약 통계 출력
print("1. 전체 특수문자 후보 요약 ")
print(df_noise2[["특수문자후보_수", "글자수", "특수문자후보_비율"]].describe().round(3))
````

### 해석
- 문서 전반에 특수문자 후보가 적지 않게 포함되어 있음
- 즉 문서 원문이 “그냥 본문 텍스트”만으로 구성된 것이 아니라, 표/번호/기호/구분선/반복 패턴이 많이 섞여 있다는 뜻
- 전처리 없이 바로 retrieval에 넣으면 일부 문서는 노이즈 영향을 받을 가능성이 높음

````python
# 특수문자 후보 개수가 많은 문서 상위 10개를 출력
print("2. 특수문자 후보가 많은 파일 상위 10개 ")
print(df_noise2[["파일명", "특수문자후보_수", "특수문자후보_비율"]].head(10))
````

### 해석
- 절대 개수가 많다는 건 문서가 길거나, 표/구조 요소가 많거나, 실제로 기호가 많이 섞여 있다는 의미
- 하지만 문서가 길수록 절대 개수는 당연히 늘 수 있으니, 이것만으로 품질이 나쁘다고 보긴 어려움

````python
# 문서 길이에 비해 특수문자 후보 비율이 높은 문서 상위 10개를 출력
print("3. 특수문자 후보 비율이 높은 파일 상위 10개 ")
print(
    df_noise2[["파일명", "특수문자후보_수", "글자수", "특수문자후보_비율"]]
    .sort_values("특수문자후보_비율", ascending=False)
    .head(10)
)
````

### 해석
- 이건 절대 개수보다 더 중요함
- 문서 길이를 감안해도 특수문자 후보 비중이 높다는 뜻이라, 실제로 텍스트 구조가 복잡하거나 cleaning 필요성이 높을 가능성이 큼
- 특히 시스템 구축/ERP/포털류 문서들이 상위권에 있는 건, 표와 번호 체계가 많은 문서 특성과도 연결될 수 있음

````python
# 자주 나오는 문자 확인

# 전체 문서에서 나온 특수문자 후보를 모을 리스트 생성
all_candidate_chars = []

# parsed 폴더 안 markdown 문서를 하나씩 순회
for fname in os.listdir(OUT_DIR):
    # markdown 파일만 분석 대상으로 사용
    if not fname.endswith(".md"):
        continue

    # 현재 파일의 전체 경로 생성
    path = os.path.join(OUT_DIR, fname)

    # 현재 문서를 읽음
    with open(path, "r", encoding="utf-8") as f:
        text = f.read()

    # 현재 문서에서 잡힌 특수문자 후보를 전체 리스트에 추가
    all_candidate_chars.extend(noise_pattern.findall(text))

# 전체 특수문자 후보의 등장 횟수를 셈
noise_counter = Counter(all_candidate_chars)

# 가장 많이 나온 후보 문자 상위 30개를 시리즈 형태로 정렬
top_noise_chars = pd.Series(noise_counter).sort_values(ascending=False).head(30)

# 결과 출력
print("4. 전체 문서에서 자주 등장하는 특수문자 후보")
print(top_noise_chars)
````

````python
# 특수문자 후보 비율이 가장 높은 문서 하나를 샘플로 선택
sample_fname = df_noise2.sort_values("특수문자후보_비율", ascending=False).iloc[0]["파일명"]

# 선택한 파일의 전체 경로 생성
sample_path = os.path.join(OUT_DIR, sample_fname)

# 샘플 문서 전체 텍스트 읽기
with open(sample_path, "r", encoding="utf-8") as f:
    sample_text = f.read()

# 샘플 문서 안에서 특수문자 후보가 등장한 위치를 모두 찾음
matches = list(noise_pattern.finditer(sample_text))

# 어떤 파일을 보는지 먼저 출력
print(f"5. 문맥 확인 대상 파일: {sample_fname}")

# 앞에서부터 20개 후보만 문맥과 함께 출력
for m in matches[:20]:
    # 후보 문자 앞쪽 25글자부터 시작
    start = max(0, m.start() - 25)

    # 후보 문자 뒤쪽 25글자까지 포함
    end = min(len(sample_text), m.end() + 25)

    # 줄바꿈은 공백으로 바꿔서 보기 쉽게 출력
    context = sample_text[start:end].replace("\n", " ")

    # 문맥 출력
    print(context)
    print("-" * 80)
````
